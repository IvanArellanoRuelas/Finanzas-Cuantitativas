# Propiedades de variables aleatorias normales 
Si yo sumo dos variables aleatorias normales podemos sumar sus varianzas y sus medias 

```python
X1 = np.random.normal(loc=5, scale=10, size=100000)
X2 = np.random.normal(loc=10, scale=20, size=100000)
X3 = np.random.normal(loc=15, scale=30, size=100000)
```

```python
print('Mean:')
print(X1.mean())
print(X2.mean())
print(X3.mean())
```

```python
X4 = X1 + X2 + X3
print(X4.mean())
```

```python
print('Var:')
print(X1.var())
print(X2.var())
print(X3.var())
```
